```python
#成人吸菸比率
#引用套件
import matplotlib.pyplot as plt
import pandas as pd
#讀取檔案
tdata=pd.read_csv("18歲以上人口目前吸菸率 (1).csv",encoding='big5')
#選定檔案內所需範圍
spots=tdata[tdata["類別"] == "整體"]
#設定折線圖配置(線條寬度、線條顏色、記號形狀、記號大小、記號顏色、記號邊框顏色)
spots[['年度','總計']].plot(kind='line',x='年度',y='總計',legend=False,linewidth=2,color="black",marker="o",ms=8,markerfacecolor="#f1404b",markeredgecolor="#a6172d")
#設定標題名稱、字體大小、粗體字
plt.title('成人吸菸比率',fontsize=15,fontweight="bold")
#設定x、y軸範圍
plt.xlim(2001,2019)
plt.ylim(0,30)
#設定x、y軸名稱、粗體字
plt.xlabel("年度",fontweight="bold")
plt.ylabel("比率(%)",fontweight="bold")
#設置用以分隔特定期間之垂直線段(範圍、粗細、顏色)
plt.vlines(x=2002, ymin=0, ymax=30,linewidth=2,color="LightSlateGray")
plt.vlines(x=2006, ymin=0, ymax=30,linewidth=2,color="LightSlateGray")
plt.vlines(x=2009, ymin=0, ymax=30,linewidth=2,color="LightSlateGray")
plt.vlines(x=2017, ymin=0, ymax=30,linewidth=2,color="LightSlateGray")
#定義x為spots中的年度範圍，y為spots中的總計範圍
x=spots["年度"]
y=spots["總計"]
#設定折線圖上的數值標示(位置、大小、顏色、粗體字)
for x,y in zip(x,y):
    plt.text(x,y+1,"%.1f"%y,ha="center",va="bottom",fontsize=8,color="#fe4365",fontweight="bold")
#標示價錢
plt.annotate("80",xy=(2001.2,9.5))
plt.annotate("85",xy=(2003.7,9.5))
plt.annotate("90",xy=(2007,9.5))
plt.annotate("100",xy=(2012,9.5))
plt.annotate("120(元)",xy=(2017.1,9.5))
#繪畫柱狀圖用來標示想強調的地方(在這邊我們想強調的是價錢)
plt.bar(2001.5,11,color="#dae9f4")
plt.bar(2001.5,9,color="white")
plt.bar(2004,11,color="#dae9f4")
plt.bar(2004,9,color="white")
plt.bar(2007.3,11,color="#dae9f4")
plt.bar(2007.3,9,color="white")
plt.bar(2012.4,11,color="#dae9f4")
plt.bar(2012.4,9,color="white")
plt.bar(2012.6,11,color="#dae9f4")
plt.bar(2012.6,9,color="white")
plt.bar(2017.6,11,color="#dae9f4")
plt.bar(2017.6,9,color="white")
plt.bar(2018.4,11,color="#dae9f4")
plt.bar(2018.4,9,color="white")
#加上格線(顏色)
plt.grid(color="#e4e7ec")
#儲存繪製好的圖
#plt.savefig('成人吸菸比率.png')
#顯示繪製好的圖
plt.show()
```

```python
#家庭二手菸
#引用套件
import pandas as pd
import matplotlib.pyplot as plt
#讀取檔案
data=pd.read_csv("C:/Users/user/Desktop/二手菸家庭暴露.csv")
#選定檔案內所需範圍
data_1=data[["年分","整體(國中)","整體(高中)"]]
#設定折線圖配置(線條寬度、線條顏色、記號形狀、記號大小、記號顏色、記號邊框顏色)
plt.plot(data_1["年分"],data_1["整體(國中)"],color="black",marker="o",ms=8,markerfacecolor="#f9320c",markeredgecolor="#a6172d")
plt.plot(data_1["年分"],data_1["整體(高中)"],color="black",marker="o",ms=8,markerfacecolor="#03a6ff",markeredgecolor="blue")
#設定線條名稱
plt.legend(labels=["國中","高中"],loc="best")
#設定x、y軸名稱、粗體字
plt.xlabel("年分",fontweight="bold")
plt.ylabel("比率(%)",fontweight="bold")
#設定標題名稱、字體大小、粗體字
plt.title("家庭二手菸暴露率",fontsize=15,fontweight="bold")
#設置用以分隔特定期間之垂直線段(範圍、粗細、顏色)
plt.vlines(x=2006, ymin=0, ymax=60,linewidth=2,color="LightSlateGray")
plt.vlines(x=2009, ymin=0, ymax=60,linewidth=2,color="LightSlateGray")
plt.vlines(x=2017, ymin=0, ymax=60,linewidth=2,color="LightSlateGray")
#設定x、y軸範圍
plt.xlim(2003,2018)
plt.ylim(15,60)
#定義x為data中的年分範圍，y為data中的整體(高中)範圍
x=data["年分"]
y=data["整體(高中)"]
#設定折線圖上的數值標示(位置、大小、顏色、粗體字)
for x,y in zip(x,y):
    plt.text(x,y-4,"%.1f"%y,ha="center",va="bottom",fontsize=8,color="#0080ff",fontweight="bold")
#定義x為data中的年分範圍，y1為data中的整體(國中)範圍
x=data["年分"]
y1=data["整體(國中)"]
#設定折線圖上的數值標示(位置、大小、顏色、粗體字)
for a,b in zip(x,y1):
    plt.text(a,b+1.5,"%.1f"%b,ha="center",va="bottom",fontsize=8,color="#fe4365",fontweight="bold")
#標示價錢
plt.annotate("85",xy=(2004.2,20.8))
plt.annotate("90",xy=(2007.2,20.8))
plt.annotate("100(元)",xy=(2012.5,20.8))
plt.annotate("120",xy=(2017.1,20.8))
#繪畫柱狀圖用來標示想強調的地方(在這邊我們想強調的是價錢)
plt.bar(2004.5,22,color="#dae9f4")
plt.bar(2004.5,20,color="white")
plt.bar(2007.5,22,color="#dae9f4")
plt.bar(2007.5,20,color="white")
plt.bar(2013,22,color="#dae9f4")
plt.bar(2013,20,color="white")
plt.bar(2013.6,22,color="#dae9f4")
plt.bar(2013.6,20,color="white")
plt.bar(2017.6,22,color="#dae9f4")
plt.bar(2017.6,20,color="white")
plt.bar(2018.4,22,color="#dae9f4")
plt.bar(2018.4,20,color="white")
#加上格線(顏色)
plt.grid(color="#e4e7ec")
#儲存繪製好的圖
#plt.savefig('家庭二手菸暴露率.png')
#顯示繪製好的圖
plt.show()
```

```python
#熱力圖，在此顯示僅為一組數據的圖
#引用套件
import folium
from folium.plugins import HeatMap
import pandas as pd
#創造一幅世界地圖
world_map = folium.Map()
#設定經緯度(這裡用的是台灣)
latitude = 25
longitude = 121
taiwan_map = folium.Map(location=[latitude, longitude], zoom_start=6)
#讀取檔案
cdata = pd.read_csv('高中職二手.csv')
cdata.head()
#在圖上顯示前21條數據
limit = 21
data = cdata.iloc[0:limit, :]
#轉換數據格式
heatdata = data[['Y','X']].values.tolist()
#將資料加到地圖內
HeatMap(heatdata).add_to(taiwan_map)
#顯示地圖
display(taiwan_map)
#儲存
#taiwan_map.save('高2015.html')

#在GitHub上添加擴展名為.ipynb的Jupyter Notebook或IPython Notebook文件時，它們將在存儲庫中呈現為靜態HTML文件。筆記本的交互式功能（例如自定義JavaScript繪圖）將無法在GitHub上的存儲庫中使用。
#所以請至：https://nbviewer.jupyter.org/github/lzhu11/108-2-PL/blob/master/%E5%B0%88%E9%A1%8C.ipynb查看地圖
```

import imageio

def create_gif(image_list, gif_name):
    frames = []
    for image_name in image_list:
        frames.append(imageio.imread(image_name))
    imageio.mimsave(gif_name, frames, 'GIF', duration = 1)
    return
